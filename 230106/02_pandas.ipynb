```python
import pandas as pd
```

```python
df = pd.read_csv("../csv/AAPL.csv")
```

```python
df.head()
```

```python
## 컬럼의 데이터를 인덱스로 변환
## set_index()
## inplace 기준의 데이터를 변경을 할 것인지 지정
df.set_index("Date", inplace=True)
```

```python
df.head(1)
```

```python
## 컬럼의 이름을 변경 
df.columns = ["시가", "고가", "저가", "종가", "수정종가", "거래량"]
```

```python
## 인덱스의 이름을 변경
df.index.name = "날짜"
```

```python
df.head(1)
```

```python
## 데이터프레임에서 특정 컬럼으로 인덱싱 
df[["시가", "종가"]]
df.loc[:, ["시가", "종가"]]
df.iloc[:, [0,3]]
```

```python
## 데이터프레임의 필터링 
## 시가가 100달러 이상인 경우
cond = df["시가"] >= 100
df.loc[cond]
```

```python
## 시가가 100 이상 150 미만인 경우 필터링
cond = (df['시가'] >= 100) & (df['시가'] < 150)
df.loc[cond]
```

```python
## query() 함수 
# query(조건식)
df.query("시가 >= 100")
```

```python
df['시가'] > df["종가"]
```

```python
df.query('시가 > 종가')
```

```python
## @뒤의 이름에 변수명을 찾아서 대입
adj_close = 150
df.query("수정종가 > @adj_close")
```

```python
## groupby
data = [
    ["2차전지(생산)", "SK이노베이션", 10.19, 1.29], 
    ["해운", "팬오션", 21.23, 1.12], 
    ["시스템반도체", "티엘아이", 35.97, 1.12], 
    ["해운", "HMM", 21.52, 3.20], 
    ["시스템반도체", "아이에이", 37.32, 3.55], 
    ["2차전지(생산)", "LG화학", 83.06, 3.75]
]
columns = ["테마", "종목명", "PER", "PBR"]
df = pd.DataFrame(data = data, columns = columns)
```

```python
df
```

```python
## 테마가 해운인 데이터랑 필터링
cond = df['테마'] == '해운'
df.loc[cond]
```

```python
df.query("테마 == '해운'")
```

```python
## 테마를 group화를 하여 평균을 구한다. 
df.groupby('테마').mean()
```

```python
df.groupby('테마').get_group('해운')
```

```python
### stack / unstack
data = [
    [100,900,800,700], 
    [1200,1400,900,800]
]
columns = [
    ['영업이익', '영업이익', '당기순이익', '당기순이익'], 
    ["컨센서스", "잠정치","컨센서스", "잠정치"]
]
df = pd.DataFrame(data = data, columns = columns)
```

```python
df
```

```python
stack_df = df.stack()
```

```python
stack2_df= stack_df.stack()
```

```python
stack2_df.unstack()
```

```python
df.stack(level=0)
```
