```python
import matplotlib.pyplot as plt
```

```python
## plot 그래프 -> 라인 그래프 생성

## 도화지에 그려질 x,y 축의 수치를 지정
plt.plot([1,2,3,4])
## 이 도화지를 그려준다.
plt.show()
```

```python
plt.plot([1,2,3,4], [1, 200, 3, 500])
plt.show()
```

```python
x = [1,2,3,4]
y = [10, 3, 4, 9]

## plot(x축의 값, y축의 값, 옵션)
## 옵션 선의 종류, 선의 색
plt.plot(x, y, "r")
plt.show()
```

```python
y2 = [4,2,5,1]
# 첫번째 라인의 값
plt.plot(x, y, "--y")
# 두번째 라인의 값
plt.plot(x, y2, ".g")
# x축의 이름
plt.xlabel("Class")
# y축의 이름
plt.ylabel("Count")
# 그래프의 title 
plt.title("Class & Count")
# 범례 추가
plt.legend(["A", "B"], loc=7)
plt.show()
```

```python
## subplot() -> 여러개의 그래프를 그릴때 영역을 나누고 영역을 지정
plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.show()
```

```python
## 바형 그래프 
x = [1,2,3,4]
y = [80, 60, 90, 70]

plt.bar(x, y)
plt.show()
```

```python
## 막대의 너비를 변경, 막대의 색 변경
plt.bar(x, y, color=["g", "r"], width=0.5)
## x축의 이름을 변경
plt.xticks(x, ["maths", "seience", "history", "english"])
plt.show()
```

```python
## 산점도 그래프
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

```python
## 산점도 그래프 옵션 
## 영역 확장, 색상 변경, 투명도 조절
color = np.random.rand(50)
area = (20 * np.random.rand(50))**2

plt.scatter(x, y, s = area, c = color, alpha = 0.6)
plt.show()
```
